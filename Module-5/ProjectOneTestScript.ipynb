```python
from CRUD_Python_Module import AnimalShelter
```

```python
shelter = AnimalShelter()
```

```python
test_record = {
    "animal_id": "CS340_TEST_001",
    "name": "ModuleFourTest",
    "animal_type": "Dog",
    "breed": "Mixed",
    "outcome_type": "Adoption"
}

create_result = shelter.create(test_record)
print("Create result:", create_result)
```

```python
results = shelter.read({"animal_id": "CS340_TEST_001"})

print("Number of records found:", len(results))
results[:1]   # display the first matching document
```

```python
update_result = shelter.update(
    {"animal_id": "CS340_TEST_001"},
    {"breed": "Labrador Mix"}
)

print("Number of records updated:", update_result)
```

```python
updated_results = shelter.read({"animal_id": "CS340_TEST_001"})
updated_results[:1]
```

```python
delete_result = shelter.delete({"animal_id": "CS340_TEST_001"})
print("Number of records deleted:", delete_result)
```

```python
post_delete_results = shelter.read({"animal_id": "CS340_TEST_001"})
print("Records found after delete:", len(post_delete_results))
```
